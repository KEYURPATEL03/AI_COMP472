```python
import numpy as np
import pandas as pd
import json 
import matplotlib.pyplot as plot
```

```python
# 1.2 loading the dataset
dataset = pd.read_json("goemotions.json") 

dataset.columns = ['posts','emotion','sentiment'] 
emotion = dataset['emotion']
sentiment = dataset['sentiment']
posts = dataset['posts']
size = dataset.shape[0]
dataset.head() #.head() to get the first n rows

#now we have a table of our dataset, lets extract the data
```

```python
#sentiments

positive = dataset.loc[sentiment == 'positive']
neutral = dataset.loc[sentiment == 'neutral']
ambiguous = dataset.loc[sentiment == 'ambiguous']
negative = dataset.loc[sentiment == 'negative']

#emotions
emotion_list = dataset['emotion'].value_counts().to_dict()
emotion_list

#calculating the percentage value of each sentiment
pos = (positive.shape[0])/size
neg = (negative.shape[0])/size
neut = (neutral.shape[0])/size
amb = (ambiguous.shape[0])/size

#calculating the count of each sentiment 
print(f'Number of positive sentiments: {positive.shape[0]}')
print(f'Number of neutral sentiments: {neutral.shape[0]}')
print(f'Number of ambiguous sentiments: {ambiguous.shape[0]}')
print(f'Number of negative sentiments: {negative.shape[0]}')
```

```python
#Plotting the distribution of sentiments as pie chart

senti_chart = np.array([pos,neg,neut,amb])
my_senti_labels = ["positive","negative","neutral","ambiguous"]
plot.pie(senti_chart)
plot.pie(senti_chart, labels = my_senti_labels)
plot.show()

#https://www.geeksforgeeks.org/matplotlib-pyplot-savefig-in-python/
plot.savefig("Sentiment_plot.pdf")
```

```python
#Plotting the distribution of emotions as pie chart

#print(emotion_list)

my_emotion_labels = [] #defining an array of labels for each emotion
emotion_size = [] #defininf an array of each emotion's size

for i, j in emotions_list.items():
    my_emotion_labels.append(i)
    emotion_size.append(j)
    
plot.pie(emotion_size, labels = my_emotion_labels)
plot.show()

plot.savefig("emotion_plot.pdf")
```
